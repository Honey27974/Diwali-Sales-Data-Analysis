```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_excel("C:/Users/Honey B Tandel/OneDrive/เอกสาร/diwali sales/Diwali Sales Data.xlsx")
sales_gen= df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending = False)

sns.barplot(x = 'Gender', y= 'Amount', data = sales_gen)
plt.show()
```

#### ***result : As you can see most of the buyers are female and even the purchasing power of female are greater than a male***

```python
ax = sns.countplot(data= df, x = 'Age Group', hue= 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df = pd.read_excel("C:/Users/Honey B Tandel/OneDrive/เอกสาร/diwali sales/Diwali Sales Data.xlsx")

sales_age= df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)

sns.barplot(x= 'Age Group', y= 'Amount', data = sales_age)
plt.show()

```

#### ***As you can see most of the buyers are of age group between 26-35***

### ***Total num of orders from top 15 state***

```python

sales_state= df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by= 'Orders', ascending= False).head(15)

sns.set(rc={'figure.figsize':(20,5)})  # give size (widht,height) so value can't overlap (see the result after comment this )
plt.xticks(rotation=-90)
sns.barplot(data= sales_state, x= 'State', y= 'Orders')
plt.show()
```

### ***total num of amount from top 10 state***

```python

sales_state= df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)

sns.set(rc={'figure.figsize':(16,5)})  # give size (widht,height) so value can't overlap (see the result after comment this )
sns.barplot(data= sales_state, x= 'State', y= 'Amount')
plt.show()

```

### ***most of the orders from uttar pradesh , maharashtra and karnataka with high paid amount but as you can see kerla are there in orders but can't see in amount.***

```python
ax = sns.countplot(data= df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_status= df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by= 'Amount',ascending= False)

sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data= sales_status, x= 'Marital_Status', y= 'Amount', hue= 'Gender')
plt.show()


```

#### ***most of the buyers are married(female) and purchasing power***

#### ***most of the buyers work on IT sector, Healthcare , Aviation***

```python
ax = sns.countplot(data= df, x = 'Product_Category')
sns.set(rc={'figure.figsize':(16,8)})
plt.xticks(rotation=-90)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Based on the amount
sales_category= df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by= 'Amount',ascending= False).head(10)
plt.xticks(fontsize=9)
sns.barplot(data= sales_category, x= 'Product_Category', y= 'Amount')
plt.show()
```

#### ***most of the sold prodcuts from food, clothing&apparel and Electronics & gadgest***

```python
# based on the orders 
sales_category= df.groupby(['Product_Category'], as_index = False)['Orders'].sum().sort_values(by= 'Orders',ascending= False).head(10)

sns.barplot(data= sales_category, x= 'Product_Category', y= 'Orders')
plt.show()
```

#### ***orders are highly come from clothing , food and electronics& gadgets***
